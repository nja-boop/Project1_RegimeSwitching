```python
data.head()
```

```python
returns = np.log(data / data.shift(1)).dropna()
returns.head()
```

```python
data.plot(figsize=(12, 6), title="Price History of SPY, TLT, GLD")
plt.ylabel("Price")
plt.xlabel("Date")
plt.grid(True)
plt.show()
```

```python
returns.plot(figsize=(12, 6), title="Daily Log Returns")
plt.ylabel("Log Return")
plt.xlabel("Date")
plt.grid(True)
plt.show()
```
